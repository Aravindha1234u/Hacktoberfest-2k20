```
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
```

```
from sklearn.datasets import load_boston
dataset = load_boston()
```

```
print(dataset.keys())
```

```
print(dataset.DESCR)
```

```
boston_df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
boston_df.head()
```

```
boston_df['MEDV'] = dataset.target
sns.set(rc={'figure.figsize':(14.5,9)})
sns.distplot(boston_df['MEDV'], bins=40)
plt.show()
```

```
correlation_matrix = boston_df.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```

```
features = ['LSTAT', 'RM']
target = boston_df['MEDV']
```

```
plt.figure(figsize=(10, 5))
x = boston_df[features[0]]
plt.scatter(x, target, marker='o')
plt.title(features[0])
plt.xlabel(features[0])
plt.ylabel('MEDV')
```

```
plt.figure(figsize=(10, 5))
x = boston_df[features[1]]
plt.scatter(x, target, marker='o')
plt.title(features[1])
plt.xlabel(features[1])
plt.ylabel('MEDV')
```
